```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('https://github.com/krishnaik06/K-Means-clustering/raw/master/Customers.csv', index_col=0)
data.head()
```

```python
X = data.iloc[:, -2:].values
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(20, 18))
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(
    n_clusters=5,
    linkage='ward'
)
y_hc = hc.fit_predict(X)
y_hc
```

```python
plt.figure(figsize=(10, 9))
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='red', label='Customers 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label='Customers 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='green', label='Customers 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='cyan', label='Customers 4')
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], s=100, c='purple', label='Customers 5')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income(K$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
nb_samples = 1000
weights = (0.95, 0.05)

X, Y = make_classification(nb_samples, weights = weights, n_features = 2, n_redundant = 0, random_state=1000)
```

```python
print(X[Y==0].shape)
print(X[Y==1].shape)
```

```python
from sklearn.utils import resample

X_1_resampled = resample(X[Y==1], n_samples=X[Y==0].shape[0], random_state = 1000)
```
